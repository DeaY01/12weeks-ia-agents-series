# Week 2 – Your First Tool-Using AI Agent
**12-Week AI Agents Series | Oladapo Ajiboye (@DeaY01) | Dec 2025**

This agent uses **real web search** to research AI agent history.
Works 100 % free — add your OpenAI key for better reasoning.

```
!pip install duckduckgo-search langchain langchain-openai -q

from langchain.tools import DuckDuckGoSearchRun
from langchain.agents import initialize_agent, AgentType
from langchain_openai import ChatOpenAI

openai_key = ""  # ← put your key here or leave empty (still works)

llm = ChatOpenAI(openai_api_key=openai_key or "dummy", temperature=0.3)

tools = [DuckDuckGoSearchRun()]

agent = initialize_agent(tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True)

print("Week 2 Tool-Using Research Agent Ready!\n")

question = "Tell me about Claude Shannon's 1950 paper and how it relates to AI agents"
print("Running example...\n")
print(agent.run(question))
```
