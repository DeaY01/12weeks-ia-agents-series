```python
!pip install openai -q

from openai import OpenAI
import textwrap

# Optional API key
api_key = ""  # ← paste your key or leave blank

client = OpenAI(api_key=api_key) if api_key.strip() else None

def history_researcher(query):
    system_prompt = (
        "You are HistoryResearcher Agent v1 — expert on AI agents 1950–2025. "
        "Provide exact dates, people, papers, and a surprising fact."
    )

    if client:
        try:
            resp = client.chat.completions.create(
                model="gpt-4o-mini",
                messages=[
                    {"role": "system", "content": system_prompt},
                    {"role": "user", "content": f"Query: {query}"}
                ],
                temperature=0.3,
                max_tokens=1200
            )
            return resp.choices[0].message["content"]
        except Exception as e:
            return f"Error: {e}\nUsing fallback history..."

    return (
        "AI Agent History Highlights (1950–2025):\n"
        "• 1950 – Claude Shannon – Early chess AI\n"
        "• 1956 – Logic Theorist – Newell & Simon\n"
        "• 1966–72 – Shakey the Robot – Stanford\n"
        "• 1987 – Subsumption Architecture – Rodney Brooks\n"
        "• 1997 – Deep Blue beats Kasparov\n"
        "• 2016 – AlphaGo\n"
        "• 2023 – Auto-GPT\n"
        "• 2024 – Devin AI\n"
        "• 2025 – African agent builders rise\n\n"
        "Surprising fact: The word 'agent' appeared in 1953 AI papers!"
    )

your_question = "Full history of AI agents 1950–2025 with key papers"
print(textwrap.fill(history_researcher(your_question), width=95))
```
